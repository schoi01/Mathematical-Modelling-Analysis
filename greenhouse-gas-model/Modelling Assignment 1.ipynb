## 3. Assumptions and Constraints

- There are sufficient $O_2$ in the atmosphere for the $CH_4$ to break down.
- The concentration of $O_2$ in the atmosphere is constant.
- All rates of change in concentration above are constant.
- The four greenhouse gases do not react with each other apart from the $CH_4$ breakdown.
- The four greenhouse gases are well mixed.
- We know the concentration of $O_2$ and it is constant.
- We know the initial concentration of the four greenhouse gases.

## 4. Building Solutions

Apply the mass balance equation to each of the four greenhouse gases:

\begin{align}
\frac{dC_0}{dt} &= R_0 + R_5  - R_6 + 2 k C_2 C_4^2\\
\frac{dC_1}{dt} &= R_1 + R_2 - R_7 - R_9 + k C_2 C_4^2\\
\frac{dC_2}{dt} &= R_3 - k C_2 C_4^2 \\
\frac{dC_3}{dt} &= R_4 - R_8  \\
C_0(0) &= C_{0,0} \quad C_1(0) = C_{1,0} \quad C_2(0) = C_{2,0} \quad C_3(0) = C_{3,0} 
\end{align}


Apply the nondimensionalization procedure. Let $C_0 = [C_0] C_0^*$, $C_1 = [C_1] C_1^*$, $C_2 = [C_2] C_2^*$, $C_3 = [C_3] C_3^*$, $C_4 = [C_4]C_4^*$, and $t = [t] t^*$. 

Make the substitutions:

\begin{align}
\frac{[C_0]}{[t]}\frac{dC_0^*}{dt^*} &= R_0 + R_5  - R_6 + [C_2][C_4]^2 2 k C_2^* C_4^{*2}\\
\frac{[C_1]}{[t]}\frac{dC_1^*}{dt^*} &= R_1 + R_2 - R_7 - R_9 + [C_2][C_4]^2 k C_2^* C_4^{*2}\\
\frac{[C_2]}{[t]}\frac{dC_2^*}{dt^*} &= R_3 - [C_2][C_4]^2 k C_2^* C_4^{*2} \\
\frac{[C_3]}{[t]}\frac{dC_3^*}{dt^*} &= R_4 - R_8  \\
[C_0]C_0^*(0) &= C_{0,0} \quad [C_1]C_1^*(0) = C_{1,0} \quad [C_2]C_2^*(0) = C_{2,0} \quad [C_3]C_3^*(0) = C_{3,0}
\quad [C_4]C_4^*(0) = C_{4,0} 
\end{align}

We choose the same scaling factor for the concentrations $[C_0] = [C_1] = [C_2] = [C_3] = [C_4] = [C]$ and simplify 
\begin{align}
\frac{dC_0^*}{dt^*} &= [t][C]^{-1} (R_0 + R_5  - R_6) + [t][C]^2 2 k C_2^* C_4^{*2} \\
\frac{dC_1^*}{dt^*} &= [t][C]^{-1} (R_1 + R_2 - R_7 - R_9) + [t][C]^2  k C_2^* C_4^{*2} \\
\frac{dC_2^*}{dt^*} &= [t][C]^{-1} R_3  - [t][C]^2 k C_2^* C_4^{*2} \\
\frac{dC_3^*}{dt^*} &= [t][C]^{-1} (R_4 - R_8)  \\
C^*_0(0) &= \frac{C_{0,0}}{[C]} \quad C^*_1(0) = \frac{C_{1,0}}{[C]} \quad C^*_2(0) = \frac{C_{2,0}}{[C]} \quad C^*_3(0) = \frac{C_{3,0}}{[C]} 
\end{align}

From the $\frac{dC^*_2}{dt^*}$ equation, we choose $[C] = (\frac{R_3}{k})^{1/3}$, and then using $[C]$ in the second term the same equation, we obtain $[t] = \frac{[C]}{R_3} = (\frac{R_3}{k})^{1/3} R_3^{-1} = k^{-1/3} R_3^{-2/3}$.

Then we can write the system of equations as

\begin{align}
\frac{dC_0^*}{dt^*} &= \mu_0  +  2  C_2^* C_4^{*2} \\
\frac{dC_1^*}{dt^*} &= \mu_1  +  C_2^* C_4^{*2} \\
\frac{dC_2^*}{dt^*} &=  1  - C_2^* C_4^{*2} \\
\frac{dC_3^*}{dt^*} &= \mu_3  \\
C^*_0(0) &= \alpha_0 \quad C^*_1(0) = \alpha_1 \quad C^*_2(0) = \alpha_2 \quad C^*_3(0) = \alpha_3
\end{align}

where $\mu_0 = \frac{(R_0 + R_5 - R_6)}{R_3} \quad \mu_1 = \frac{(R_1 + R_2 - R_7 - R_9)}{R_3} \quad \mu_3 = \frac{(R_4 - R_8) }{R_3} \quad \alpha_0 = (\frac{k}{R_3})^{\frac{1}{3}}C_{0,0}$
$\quad \quad \alpha_1 =(\frac{k}{R_3})^{\frac{1}{3}}C_{1,0} \quad \alpha_2 = (\frac{k}{R_3})^{\frac{1}{3}}C_{2,0} \quad \alpha_3 = (\frac{k}{R_3})^{\frac{1}{3}}C_{3,0}.$

## 5. Analyze and Assess

```python
# constants
mu_0 = 0.001
mu_1 = 0.2    
mu_3 = 0.049
c4 = 1.5        # oxygen constant
    
# initial conditions
c0 = 100        # initial H2O
c1 = 100        # initial CO2
c2 = 100        # initial CH4
c3 = 100        # initial N2O

# DE function
def f(C,t):
    dCdt = np.array([0.,0.,0.,0.,])
    dCdt[0] = mu_0 + 2*C[2]*c4**2
    dCdt[1] = mu_1 + C[2]*c4**2
    dCdt[2] = 1 - C[2]*c4**2
    dCdt[3] = mu_3
    return dCdt

# initial condition
C0 = [c0,c1,c2,c3]
# time vector
t = np.linspace(0,1000,10000)  # long term, change values as needed
# DE solution
C = spi.odeint(f,C0,t)

#plot
## uncomment the following lines of code to show graph

# plt.plot(t,C[:,0],t,C[:,1],t,C[:,2],t,C[:,3])
# plt.grid(True)
# plt.legend(['c0 $H_2O$','c1 $CO_2$','c2 $CH_4$','c3 $N_2O$'])
# plt.xlabel('$t^*$')
# plt.ylabel('$C^*$')
# plt.show()
```

### Observations and Analysis

The graph on the left in Figure 1 represents the baseline behaviour of each gas concentration. We set the net concentration for all four greenhouse gases equal to $0$ while keeping a  constant nonzero oxygen ($O_2$; $c_4$) concentration such that the breakdown of methane could still happen. We see that methane ($CH_4$) and nitrous oxide ($N_2O$) concentration have a steady state close to $0$, while water vapour ($H_2O$) and carbon dioxide ($CO_2$) show a linearly increasing trend. When we increased the initial condition value of $O_2$ and $CH_4$ concentration ($c_2$; green line), the concentration for both $H_2O$ and $CO_2$ increased more rapidly in the beginning. We observed that the concentration of $CH_4$ reached a lower steady state when the initial values of concentration of $CH_4$ and $O_2$ were higher. These findings indicate that $CH_4$ (combined with $O_2$) breaks down into $CO_2$ and $H_2O$. As a result, when more $CH_4$ and $O_2$ are present in the atmosphere, the breakdown happens more rapidly leading to a . higher concentration of $CO_2$ and $H_2O$.





In the model, each $\mu$ represents the net balance between emission and absorption of the gases. Changing these $\mu$ values ($\mu_0, \mu_1, \mu_3$) corresponding to $H_2O$, $CO_2$, and $N_2O$, respectively, seemed to affect the slope of their concentration curves. For instance, larger $\mu$ values produced a steeper linear graph, suggesting a stronger net source over time. When plotting many different graphs with alternated initial conditions, we observed that different initial conditions changed the starting values of the lines, while the slope remained the same. This aligns with our deterministic model, because $\frac{dC_3}{dt} = \mu_3$ is constant, allowing its slope to solely depend on $\mu_3$.

The concentration rates for $CH_4$ and $N_2O$ appeared to be less susceptible to change compared to those for $H_2O$ and $CO_2$. Again, this observation aligns with our model, as the rates of $CH_4$ and $N_2O$ depend only on constant parameters ($c_4$ and $\mu_3$) so they are less sensitive to other sources (i.e., evaporation and precipitation) compared to $H_2O$ and $CO_2$.

The graph on the left of Figure 2 shows a short term behaviour ($t \in [0,10]$), while that on the right shows a long term behaviour ($t \in [0,1000]$). When we looked at these curves in the short term, they all seemed to come to certain steady states after an initial curve (either a decay or growth). However, when looked at in the long run, we observed that the three gases ($H_2O$, $CO_2$, and $N_2O$) still increased linearly, with $CH_4$ still coming to a steady state of $0$.



### Model Limitations

Since we assumed the volcanic eruption released the greenhouse gases at a constant rate, the lines of $H_2O$ and $CO_2$ in the plots looked smoother than the real case. We also assumed that all rates of emission and absorption of greenhouse gases were at constant rates, which erased the fluctuation of the concentration lines. Furthermore, the sources that emit greenhouse gases might not keep emitting gases forever, so our model only gives information for the concentrations in the foreseeable future.
